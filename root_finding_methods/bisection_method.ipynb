```python
from typing import Callable
from math import sin, log, exp

def bisection_method(
        f: Callable[[float], float],
        a : float, 
        b : float,
        tolerance : float) -> float:

    if a == b and (not abs(f(a)) < tolerance):
        return None

    a, b = (a, b) if a < b else (b, a)

    guesses : list[tuple[float, float]] = []

    guess = bisect(a, b)
    while abs(f(guess)) > tolerance:
        guesses.append((guess, f(guess)))
        if f(guess) > 0:
            b = guess
        elif f(guess) < 0:
            a = guess
        guess = bisect(a, b)
            

    return guesses


def bisect(a : float, b : float) -> float:
    return a + ((b - a) / 2)

import matplotlib.pyplot as plt

func_pow = lambda x : ( x ** 2) - 2
func_sin = sin
func_log = log

guesses_pow = bisection_method(func_pow, 0, 2, 0.005)
guesses_sin = bisection_method(func_sin, 0, 2, 0.005)
guesses_log = bisection_method(func_log, 1, 2, 0.005)

#plt.scatter(zip(*guesses_exp))
x, y = zip(*guesses_pow)
plt.xlabel('x')
plt.ylabel('(x^2) - 2')
plt.scatter(x, y, c=x, cmap='viridis')
plt.show()
print("closest pair " + str(guesses_pow[-1][0]) + ", " + str(guesses_pow[-1][1]) + ")")

x, y = zip(*guesses_sin)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.scatter(x, y, c=x, cmap='viridis')
plt.show()
print("closest pair " + str(guesses_sin[-1][0]) + ", " + str(guesses_sin[-1][1]) + ")")

x, y = zip(*guesses_log)
plt.xlabel('x')
plt.ylabel('log(x)')
plt.scatter(x, y, c=x, cmap='viridis')
plt.show()
print("closest pair " + str(guesses_log[-1][0]) + ", " + str(guesses_log[-1][1]) + ")")


```
